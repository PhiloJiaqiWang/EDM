```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
tutor = pd.read_csv('CognitiveTutorAlgebra-gaming-clips.csv')
tutor
```

## Question 1

Compute each of the following 3 features:
AvgTimePerAction: The average time, in seconds, spent on each action included in the clip.
ContextChanges: The number of times the context (element of the user interface) in which the action took place was different from the previous actions for each action of the clip.
CorrectActions: The number of times the student executed a correct action within the clip.

```python
tutor = tutor.dropna()
shifted_tutor = tutor.shift(periods=1, fill_value=np.nan)
tutor['Changed'] = ((tutor["context"] != shifted_tutor["context"]) & (tutor['Gaming clip'] == shifted_tutor['Gaming clip'])).astype(int)
tutor["ifRight"] = np.where(tutor['assessment'] == 'RIGHT', 1, 0)
tutor
```

```python
AvgTimePerAction= tutor.groupby("Gaming clip")["time"].mean().reset_index()
ContextChanges = tutor.groupby("Gaming clip")["Changed"].sum().reset_index()
CorrectActions = tutor.groupby("Gaming clip")["ifRight"].sum().reset_index()
clip_tutor = pd.merge(AvgTimePerAction, ContextChanges, on='Gaming clip', suffixes=('AvgTimePerAction', 'ContextChanges'))
clip_tutor = pd.merge(clip_tutor,CorrectActions, on='Gaming clip', suffixes=('AvgTimePerAction', 'ContextChanges','CorrectActions'))
clip_tutor = clip_tutor.rename(columns = {'time':'AvgTimePerAction','Changed': 'ContextChanges', 'ifRight': 'CorrectActions'})
clip_tutor
```

## Question 2

Develop and compute 2 features that might be useful when trying to identify the gaming the system behavior. Provide a meaningful name and a description for both features you created.

### 1. HintRate

This feature is calculated by hint_count/action_count, it shows how often the student is trying to get hint, the higher it is, the more chance they are trying to game the system.

```python
action_num= tutor.groupby("Gaming clip")["action"].count().reset_index()
tutor["ifHint"] = np.where(tutor['assessment'] == 'HELP', 1, 0)
hint_num= tutor.groupby("Gaming clip")["ifHint"].sum().reset_index()
clip_hint = pd.merge(action_num, hint_num, on='Gaming clip')
clip_hint["HintRate"]= clip_hint["ifHint"]/clip_hint["action"]
clip_tutor = pd.merge(clip_tutor,clip_hint, on='Gaming clip')
clip_tutor  = clip_tutor.drop(columns=["action", "ifHint"])
clip_tutor 
```

### 2.BugRate

This feature is calculated by bug_count/action_count, it shows how often the student is inputting something can't be recoginized by the system, it is an indication of potential guessing. The higher it is, the more likely the student is trying to guessing the answer.

```python
tutor["ifBug"] = np.where(tutor['assessment'] == 'BUG', 1, 0)
bug_num= tutor.groupby("Gaming clip")["ifBug"].sum().reset_index()
clip_bug = pd.merge(action_num, bug_num, on='Gaming clip')
clip_bug["BugRate"]= clip_bug["ifBug"]/clip_bug["action"]
clip_tutor = pd.merge(clip_tutor,clip_bug, on='Gaming clip')
clip_tutor  = clip_tutor.drop(columns=["action", "ifBug"])
clip_tutor 
```
