### 2. Function Defining

```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import RFE
from sklearn.model_selection import StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import cohen_kappa_score

# Define your features (X) and target (y)
X = gaming.drop('is gaming?', axis=1)  # Replace 'target_column' with the actual target column name
y = gaming['is gaming?']
X = X.drop(['student', 'gaming clip'], axis = 1)
batch = gaming['batch']

def ApplyModel(model_type, num):
    # Specify the number of folds for cross-validation
    num_folds = 5  # You can adjust the number of folds as needed

    # Initialize the StratifiedKFold cross-validation with 'batch' as the grouping variable
    stratified_kfold = StratifiedKFold(n_splits=num_folds, shuffle=True, random_state=42)

    # Create an instance of the classification model (e.g., Logistic Regression)
    model = model_type

    # Initialize RFE for feature selection
    rfe = RFE(estimator=model, n_features_to_select=num)  # Specify the number of features to select

    # Initialize a list to store cross-validation results
    kappa_scores = []

    # Perform cross-validation with stratification by 'batch' and feature selection
    for train_index, test_index in stratified_kfold.split(X, batch):
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]
        
        # Fit RFE on the training data
        rfe.fit(X_train, y_train)

        # Select the features
        selected_features = X_train.columns[rfe.support_]
        X_train_selected = X_train[selected_features]
        X_test_selected = X_test[selected_features]
        
        if "batch" in selected_features:
            selected_features = [feature for feature in selected_features if feature != "batch"]

        # Fit the model on the training data with selected features
        model.fit(X_train_selected, y_train)

        # Make predictions on the test data
        y_pred = model.predict(X_test_selected)

        # Calculate Cohen's Kappa for this fold
        kappa = cohen_kappa_score(y_test, y_pred)
        kappa_scores.append(kappa)

    # Calculate the mean Cohen's Kappa across all folds
    mean_kappa = np.mean(kappa_scores)
    return mean_kappa, selected_features
```

### 3. Logistic Regression

```python
model = LogisticRegression()
mean_kappa, selected_features = ApplyModel(model, 5)
```

### 6. Neural Net

```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import cohen_kappa_score
from sklearn.preprocessing import MinMaxScaler  # Import Min-Max Scaler
import tensorflow as tf
from tensorflow import keras
from sklearn.preprocessing import LabelEncoder


# Define the neural network model
def create_neural_network(input_dim):
    model = keras.Sequential([
        keras.layers.Input(shape=(input_dim,)),
        keras.layers.Dense(64, activation='relu'),
        keras.layers.Dense(32, activation='relu'),
        keras.layers.Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

def neural(X, y, batch, num_folds=5):
    scaler = MinMaxScaler()  # Use Min-Max Scaler
    X = scaler.fit_transform(X)
    label_encoder = LabelEncoder()
    y = label_encoder.fit_transform(y)

    stratified_kfold = StratifiedKFold(n_splits=num_folds, shuffle=True, random_state=42)

    kappa_scores = []

    k = 5

    selector = SelectKBest(chi2, k=k)
    

    for train_index, test_index in stratified_kfold.split(X, batch):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]

        X_train_selected = selector.fit_transform(X_train, y_train)
        X_test_selected = selector.transform(X_test)
        
        selected_feature_indices = selector.get_support(indices=True)

        model = create_neural_network(X_train_selected.shape[1])
        model.fit(X_train_selected, y_train, epochs=10, batch_size=32, verbose=0)

        y_pred = (model.predict(X_test_selected) > 0.5).astype(int)

        kappa = cohen_kappa_score(y_test, y_pred)
        kappa_scores.append(kappa)

    mean_kappa = np.mean(kappa_scores)

    return mean_kappa, selected_feature_indices
```

```python
game_fixed = gaming.drop(['student', 'gaming clip','is gaming?'], axis = 1)
mean_kappa, selected_feature_indices = neural(game_fixed, gaming['is gaming?'], gaming['batch'])
print("Mean Cohen's Kappa:", mean_kappa)
print("selected_feature_indices:", X.columns[selected_feature_indices])
```

### 7. LSTM

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

def lstm(X, y, batch, num_folds=5):
    scaler = MinMaxScaler()  # Use Min-Max Scaler
    X = scaler.fit_transform(X)
    label_encoder = LabelEncoder()
    y = label_encoder.fit_transform(y)

    stratified_kfold = StratifiedKFold(n_splits=num_folds, shuffle=True, random_state=42)

    kappa_scores = []

    k = 5

    selector = SelectKBest(chi2, k=k)

    selected_feature_indices_list = []  # To store selected feature indices for each fold

    for train_index, test_index in stratified_kfold.split(X, batch):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]

        X_train_selected = selector.fit_transform(X_train, y_train)
        X_test_selected = selector.transform(X_test)

        selected_feature_indices = selector.get_support(indices=True)
        selected_feature_indices_list.append(selected_feature_indices)

        # Reshape X_train and X_test to include sequence length dimension
        sequence_length = 1  # Since we are not dealing with sequential data
        X_train_selected_reshaped = X_train_selected.reshape(X_train_selected.shape[0], sequence_length, -1)
        X_test_selected_reshaped = X_test_selected.reshape(X_test_selected.shape[0], sequence_length, -1)

        # Define and compile the LSTM model
        model = Sequential()
        model.add(LSTM(64, input_shape=(sequence_length, X_train_selected.shape[1]), activation='relu'))
        model.add(Dense(1, activation='sigmoid'))
        model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

        # Train the model
        model.fit(X_train_selected_reshaped, y_train, epochs=10, batch_size=32, verbose=0)

        # Predict and evaluate
        y_pred = (model.predict(X_test_selected_reshaped) > 0.5).astype(int)

        kappa = cohen_kappa_score(y_test, y_pred)
        kappa_scores.append(kappa)

    mean_kappa = np.mean(kappa_scores)

    return mean_kappa, selected_feature_indices_list
```

```python
game_fixed = gaming.drop(['student', 'gaming clip','is gaming?'], axis = 1)
mean_kappa, selected_feature_indices = lstm(game_fixed, gaming['is gaming?'], gaming['batch'])
print("Mean Cohen's Kappa:", mean_kappa)
print("selected_feature_indices:", X.columns[selected_feature_indices])
```
